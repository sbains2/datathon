```python
import pandas as pd
df = pd.read_csv('us_canada_border.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%b %Y')
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Drop the original 'Date' column if no longer needed
df = df.drop(columns=['Date'])
descriptive_stats = df.groupby(['Port Name', 'Year'])['Value'].describe()
descriptive_stats
```

```python

total_avgs = df.groupby('Year').describe()
total_avgs['Value']
```

```python
state_avgs = df.groupby(['State', 'Year']).describe()
state_avgs['Value']
```
